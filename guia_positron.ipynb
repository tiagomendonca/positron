## Guia para o Positron

## Atalhos

### Positron

| Ação                                       | Atalho (sequência)                   |
|--------------------------------------------|--------------------------------------|
| Mostra todos os comandos                   | Ctrl + Shift + P                     |
| Mostra os atalhos do teclado               | Shift + Alt + K                      |
| Executa código                             | Ctrl + Enter                         |
| Executa código desde o início até o cursor | Ctrl + Alt + B                       |
| Salvar                                     | Ctrl + S                             |
| Operador pipe %>% ou \|>                   | Ctrl + Shift + M                     |

### Notebook

| Ação                                | Atalho (sequência)                   |
|-------------------------------------|--------------------------------------|
| Executa célula                      | Ctrl + Alt + Enter ou Ctrl + Enter   |
| Executa célula e seleciona abaixo   | Shift + Enter                        |
| Muda tipo da célula para código     | Y                                    |
| Muda tipo da célula para markdown   | M                                    |
| Adiciona célula acima               | A                                    |
| Adiciona célula abaixo              | B                                    |
| Move célula para cima               | Alta + seta para cima                |
| Move célula para baixo              | Alta + seta para baixo               |
| Apaga célula                        | D D                                  |
| Divide célula                       | Ctrl + Shift + -                     |


### Assistentes

No momento, é possível utilizar os seguintes modelos no Positron.


| Provider | Chat | Code Completions | Authentication | 
| :--- | :---: | :---: | :--- | 
| Anthropic | ok | &nbsp;| API Key | 
| GitHub Copilot (Preview) | ok | ok | GitHub OAuth |
| Amazon Bedrock (Preview) | ok | &nbsp;  | AWS CLI or Posit Workbench Managed Credentials |
| Snowflake Cortex (Preview) | ok | &nbsp; | API Key or Posit Workbench Managed Credentials |
| OpenAI (Preview) | ok | &nbsp; | API Key |
| Custom Provider (Experimental) | ok |&nbsp; | API Key |


Para configurar os assistentes, acesse [https://positron.posit.co/assistant-getting-started.html](https://positron.posit.co/assistant-getting-started.html)

Sobre utilizar OpenAI de assistente: OpenAI provider support is currently in preview and must be manually enabled.
Run the command Preferences: Open User Settings (JSON) and add the following to your settings file:

"positron.assistant.enabledProviders": ["openai-api"]

Isso vai abrir o seguinte campo para adicionar o modelo da OpenAI.
